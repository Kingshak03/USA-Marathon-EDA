### Comparing Gender Distribution Across Race Lengths

The following plot, generated using `sns.histplot(df, x='race_length', hue='athlete_gender')`, compares male and female participation across the two race lengths: **50km** and **50 miles**.

#### Observations:
- In the **50km race**, there is a more balanced distribution and an even ratio of male to female athletes.
- In contrast, the **50-mile race** shows a less even distribution, with one gender (male) dominating participation.

This highlights that shorter race lengths attract a more diverse set of participants in terms of gender compared to longer races.

### Average Speed by Gender and Race Length

The table below summarizes the average speeds of athletes grouped by race length and gender. The data provides insights into performance differences across the 50km and 50-mile races for male and female participants.

#### Observations:
- In the **50km race**:
  - Female athletes have an average speed of **7.08 units**.
  - Male athletes have a higher average speed of **7.74 units**, showing a more pronounced difference between genders.
- In the **50-mile race**:
  - Female athletes have an average speed of **6.83 units**.
  - Male athletes have an average speed of **7.26 units**, but the difference between genders is much closer compared to the 50km race.

Overall, while male participants consistently exhibit higher average speeds than female participants, the gap in performance is notably smaller in the longer 50-mile race.

```python
# Seasons for the data -> Are runners slower in summer than winter?

# Spring 3-5
# Summer 6 - 8
# Fall 9 - 11
# Winter 12 - 2

# Split between two decimals, using race day
```

```python
# European Date
df['race_month'] = df['race_day'].str.split('.').str.get(1).astype(int)
df.tail(2)
```

```python
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'

df['race_season'] = df['race_month'].apply(get_season)
df.head(2)
```

```python
df.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# 50 Miler Only
df.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
